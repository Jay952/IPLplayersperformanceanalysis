```python
runs = 0
balls = 0
dismissals = 0

for i in DavidWarner['batsman_runs']:
    runs = runs + i
    balls = balls + 1

for j in DavidWarner['is_wicket']:
    if j == 1:
        dismissals = dismissals + 1
        
print("Total runs: ", runs)
print("Total balls faced: ", balls)
print("Number of dismissals: ", dismissals)
```

### David Warner innings segments

```python
unique_match_id=[]
no_of_balls=[]
current_balls=0
for i in DavidWarner['id']:
    if i not in unique_match_id:
        if(current_balls>0):
            no_of_balls.append(current_balls)
        current_balls=1
        unique_match_id.append(i)
    else:
        current_balls=current_balls+1 
no_of_balls.append(current_balls)
```

```python
no_of_balls
```

```python
runs_0_to_15 = 0
runs_15_to_30 = 0
runs_30_above = 0

balls_0_to_15 = 0
balls_15_to_30 = 0
balls_30_above = 0

dismissals_0_to_15 = 0
dismissals_15_to_30 = 0
dismissals_30_above = 0
```

```python
index=0
ball=0
for i in DavidWarner['batsman_runs']:
    ball=ball+1
    if(ball<15):
        runs_0_to_15 = runs_0_to_15 + i
        balls_0_to_15 = balls_0_to_15 + 1
    elif(ball>=15 and ball<=30):
        runs_15_to_30 = runs_15_to_30 + i
        balls_15_to_30 = balls_15_to_30 + 1
    elif(ball>30):
        runs_30_above = runs_30_above + i
        balls_30_above = balls_30_above + 1
    if(ball==no_of_balls[index]):
      
        index=index+1
        ball=0    
```

```python
index=0
ball=0
for i in DavidWarner['is_wicket']:
    ball=ball+1
    if(ball<15):
        if(i==1):
            dismissals_0_to_15 = dismissals_0_to_15 + 1
    elif(ball>=15 and ball<=30):
        if(i==1):
            dismissals_15_to_30 = dismissals_15_to_30 + 1
    elif(ball>30):
        if(i==1):
            dismissals_30_above = dismissals_30_above + 1

    if(ball==no_of_balls[index]):
      
        index=index+1
        ball=0   
```

```python
sr_0_to_15 = runs_0_to_15/balls_0_to_15*100
avg_0_to_15 = runs_0_to_15/dismissals_0_to_15

print("Strike rate in 0-15 balls: ", sr_0_to_15)
print("Average in 0-15 balls: ", avg_0_to_15)
```

```python
sr_15_to_30 = runs_15_to_30*100/balls_15_to_30
avg_15_to_30 = runs_15_to_30/dismissals_15_to_30

print("Strike rate in 15-30 balls: ", sr_15_to_30)
print("Average in 15-30 balls: ", avg_15_to_30)
```
